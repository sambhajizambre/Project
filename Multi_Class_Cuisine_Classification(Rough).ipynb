```python

def cleanPunc(sentence): 
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned
```

```python
new_df.columns = [col.replace('[^a-zA-Z#', '') for col in new_df.columns]
# Removing 'Punctuation' from column names
```

```python
new_df
```

**Model Building**

```python
seeds = [1, 43, 678, 90, 135]
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df['detailed_address'], 
                                                    new_df[new_df.columns[1:]], 
                                                    test_size=0.3, 
                                                    random_state=seeds[4], 
                                                    shuffle=True)
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
vectorizer.fit(X_train)

X_train = vectorizer.transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
```

```python
LR_pipeline = Pipeline([('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1))])
accuracy = 0
for detailed_address in new_df.columns[1:]:
    print('**Processing {} titles...**'.format(detailed_address))
    
    # Training logistic regression model on train data
    LR_pipeline.fit(X_train, y_train[detailed_address])
    
    # calculating test accuracy
    prediction = LR_pipeline.predict(X_test)
    accuracy = accuracy + accuracy_score(y_test[detailed_address], prediction)
    print('AUC ROC score is {}'.format(roc_auc_score(y_test[detailed_address],prediction)))
    print("\n")

print('Test averaged AUC ROC is {}'.format(accuracy/len(new_df.columns[1:])))
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_pipeline = Pipeline([('clf', OneVsRestClassifier(MultinomialNB(
                    fit_prior=True, class_prior=None)))])
accuracy = 0
for detailed_address in new_df.columns[1:]:
    print('**Processing {} titles...**'.format(detailed_address))
    
    # Training naive_bayes model on train data
    NB_pipeline.fit(X_train, y_train[detailed_address])
    
    # calculating test accuracy
    prediction = NB_pipeline.predict(X_test)
    accuracy = accuracy + accuracy_score(y_test[detailed_address], prediction)
    print('AUC ROC is {}'.format(roc_auc_score(y_test[detailed_address],prediction)))
    print("\n")

print('Test averaged AUC ROC is {}'.format(accuracy/len(new_df.columns[1:])))
```

```python
from sklearn.svm import LinearSVC

SVC_pipeline = Pipeline([('clf', OneVsRestClassifier(LinearSVC(), n_jobs=1))])

accuracy = 0
for detailed_address in new_df.columns[1:]:
    print('**Processing {} titles...**'.format(detailed_address))
    
    # Training linearSVC model on train data
    SVC_pipeline.fit(X_train, y_train[detailed_address])
    
    # calculating test accuracy
    prediction = SVC_pipeline.predict(X_test)
    accuracy = accuracy + roc_auc_score(y_test[detailed_address], prediction)
    print('F1-score is {}'.format(roc_auc_score(y_test[detailed_address],prediction)))
    print("\n")

print('Test averaged f1-score is {}'.format(accuracy/len(new_df.columns[1:])))
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier

RF_pipeline = Pipeline([('clf', OneVsRestClassifier(RandomForestClassifier(), n_jobs=1))])

accuracy = 0
for detailed_address in new_df.columns[1:]:
    print('**Processing {} titles...**'.format(detailed_address))
    
    # Training Random Forest model on train data
    RF_pipeline.fit(X_train, y_train[detailed_address])
    
    # calculating test accuracy
    prediction = RF_pipeline.predict(X_test)
    accuracy = accuracy + roc_auc_score(y_test[detailed_address], prediction)
    print('AUC ROC is {}'.format(roc_auc_score(y_test[detailed_address],prediction)))
    print("\n")

print('Test averaged AUC ROC is {}'.format(accuracy/len(new_df.columns[1:])))
```
