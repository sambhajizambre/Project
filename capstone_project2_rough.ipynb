```python
df=df.drop('listed_in(type)',axis=1)
```

```python
n = df.nunique(axis=0)
n
```

```python
n = df.groupby('listed_in(city)')
n.first()
n.describe()
```

```python
df['listed_in(city)'] = df['listed_in(city)'].fillna('Electronic City')
null=df.isna().sum()
null

```

```python
n = df.groupby('approx_cost(for two people)')
n.first()
n.describe()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(800)
null=df.isna().sum()
null
```

```python
n = df.groupby('dish_liked')
n.first()
n.describe()
```

```python
df['dish_liked'] = df['dish_liked'].fillna(1)
null=df.isna().sum()
null
```

```python
n = df.groupby('rate')
n.first()
n.describe()
```

```python
df['rate'] = df['rate'].fillna(4.7)
null=df.isna().sum()
null
```

```python
df.head(2)
```

```python
df['rate'] = df['rate'].str.split('/').str[0]
df.loc[df['rate']=="NEW", 'rate'] = np.nan
df.loc[df['rate']=="-", 'rate'] = np.nan
df['rate'] =df['rate'].astype('float')
```

```python
location_rating=df.groupby(['location'])['rate'].mean().sort_values()[-50:][:-1]
location_rating
```

```python
location_rating.describe()
```
